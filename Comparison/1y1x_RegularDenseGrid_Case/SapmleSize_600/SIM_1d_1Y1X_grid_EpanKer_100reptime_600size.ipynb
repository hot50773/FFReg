```python

'''
---------------------------------------------------------------
Simulation for Function-on-Function Linear Regression
---------------------------------------------------------------
* 1 d

* 1 Response Y & 1 Predictor X   

* Regular Dense Grid (num of grid: 41)

* Training samp size: 600
  Test     samp size: 300

* Using Epan Kernal Function in LPR

* Using Leave-One-Curve-Out CV for bandwidth selection in LPR 

* Using FVE
---------------------------------------------------------------
'''
#####
import os
os.chdir('/home/hot50773')
import GetData
import FFReg
import Plotting
import lpr
import time
import scipy as sp
from scipy import io
import numpy as np
os.chdir('/home/hot50773/SIM_1y1x_grid1_size600')

# ---------
# Settings
# ---------

tran_samp_size = 600
test_samp_size = 300
samp_size = tran_samp_size + test_samp_size

num_pts = 41

# -------------
# Setting of X
# -------------
def X_Mean_Func(s):
    #input, output: vector
    return(s /  + 2 * np.sin(s))

def Eigen_Funcs_X(s):
    #input  :d-vector
    #output:array with dim (2 * d)
    return(np.array([np.sqrt(2 / 5) * -np.cos(2 * np.pi * s / 5), 
                     np.sqrt(2 / 5) *  np.sin(2 * np.pi * s / 5),
                     np.sqrt(2 / 5) * -np.cos(4 * np.pi * s / 5)]))

num_grid_X = num_pts
range_X = np.array([0, 5])
X_eigen_val = np.array([9, 4, 1])

# -------------
# Setting of Y
# -------------

def Y_Mean_Func(s):
    #input, output: vector
    return(s / 2 + np.sin(s))

def Eigen_Funcs_Y(s):
    #input  :d-vector
    #output:array with dim (2 * d)
    return(np.array([np.sqrt(2 / 5) * -np.cos(np.pi * s / 5), 
                     np.sqrt(2 / 5) * np.sin(np.pi * s / 5),
                     np.sqrt(2 / 5) * -np.cos(2 * np.pi * s / 5)])) 

num_grid_Y = num_pts

range_Y = np.array([0, 5])

B = np.array([[1, 0.8, 0.8], [-1, 0.5, -0.5],[0.8, -0.5, 0.2]])
```

```python
X_time_grid = sp.io.loadmat('/home/hot50773/SIM_1y1x_grid1_size600/sim_X_time_grid.mat')['grid'][0]
Y_time_grid = sp.io.loadmat('/home/hot50773/SIM_1y1x_grid1_size600/sim_Y_time_grid.mat')['grid'][0]
```

```python
# bandwidth
X_candidate_h_mean = np.arange(0.25, 1.25, 0.05).reshape(-1, 1)
X_candidate_h_cov = np.arange(0.25, 1.25, 0.05).repeat(2).reshape(-1, 2)
X_candidate_h_diag_cov = np.arange(0.25, 1.25, 0.1).reshape(-1, 1)

Y_candidate_h_mean = np.arange(0.25, 1.25, 0.05).reshape(-1, 1)
Y_candidate_h_cov = np.arange(0.25, 1.25, 0.05).repeat(2).reshape(-1, 2)

X_Y_candidate_h_cov = np.asanyarray(np.meshgrid(np.arange(0.3, 0.7, 0.05), 
                                                 np.arange(0.3, 0.7, 0.05))).T.reshape(-1,2)
```

```python
sim_rep_time = 100
```

```python
#Simulation

my_MISE   = np.zeros(sim_rep_time)
# my_RMSPE  = np.zeros(sim_rep_time)
# my_MaxAbs = np.zeros(sim_rep_time)
X_data = []
Y_data = []

for i in range(sim_rep_time):
    start_time = time.time()
    #load data from mat
    X_sim_data_i = sp.io.loadmat('/home/hot50773/SIM_1y1x_grid1_size600/sim_X_data_'+str(i+1)+'.mat')
    Y_sim_data_i = sp.io.loadmat('/home/hot50773/SIM_1y1x_grid1_size600/sim_Y_data_'+str(i+1)+'.mat')
    
    #convert into our data form
    X_data.append(GetData.Load_Data(sim_data = X_sim_data_i,time_grid=X_time_grid, tran_test_size = np.array([tran_samp_size, test_samp_size])))
    Y_data.append(GetData.Load_Data(sim_data = Y_sim_data_i,time_grid=Y_time_grid, response = True,tran_test_size = np.array([tran_samp_size, test_samp_size])))
    
    '''
    Results of X
    '''

    result_X = FFReg.Get_FPCA_Result(X_data[i], X_candidate_h_mean, 
                                      X_candidate_h_cov, X_candidate_h_diag_cov, 
                                      ker_fun = 'Epan', fve = 0.98)
    #fitting test sample
    fit_test_X = result_X.Restruect_Fun(X_data[i].test_time_pts.reshape(test_samp_size, num_pts, 1),
                                         X_data[i].obs_test)[1]

    '''
    Results of Y
    '''


    result_Y = FFReg.Fit_Mean_and_Cov(Y_data[i], Y_candidate_h_mean, 
                                      Y_candidate_h_cov, ker_fun = 'Epan')


    '''
    Construct Coefficient Function

        1. fit cov_Xi_Xj
        2. fit cov_Xi_Y
        3. combine them to build coefficient function

    '''



    fit_cov_XY = FFReg.Fit_Cov_XY(X_time_pts = X_data[i].tran_time_pts,
                                  Y_time_pts = Y_data[i].tran_time_pts,
                                  obs_X = X_data[i].obs_tran, 
                                  obs_Y = Y_data[i].obs_tran,
                                  X_time_grid = X_data[i].time_grid,
                                  Y_time_grid = Y_data[i].time_grid, 
                                  fit_X_mean = result_X.mean_fun, 
                                  fit_Y_mean = result_Y.mean_fun, 
                                  candidate_h_cov = X_Y_candidate_h_cov, 
                                  ker_fun = 'Epan')

    '''
    Coefficient Function
    '''

    fit_cov_XX = result_X.cov_fun

    inv_fit_cov_XX = np.linalg.pinv(fit_cov_XX, rcond= 0.07)

    fit_Beta_without_delta =  np.matmul(inv_fit_cov_XX,fit_cov_XY)

    ## add delta
    if i == 0:
        delta = np.diag(np.repeat([X_data[i].time_grid[1] - X_data[i].time_grid[0]],
                                   [num_grid_X]))
        delta_inv = np.linalg.inv(delta)

    fit_Beta  = np.matmul(delta_inv, fit_Beta_without_delta)


    #---------------------------------------
    # fit Response of test set on grid
    #---------------------------------------
    fit_test_X_center = fit_test_X - result_X.mean_fun


    #compute test_Y(on grid pts)
    fit_test_Y = result_Y.mean_fun + np.matmul(fit_Beta_without_delta.T, fit_test_X_center.T).T


    #real value of test_Y (on grid pts)
    real_test_Y = Y_data[i].real_val_on_grid[tran_samp_size:,:]

    #========================================================
    #compute RMSPE (relative mean squared prediction error)
    #========================================================
    grid_width = Y_data[i].time_grid[1] - Y_data[i].time_grid[0]
#     my_RMSPE[i] = (np.sum(np.square(real_test_Y - fit_test_Y) * grid_width, 1) /np.sum((real_test_Y**2) * grid_width, 1)).mean()
    #======================================
    #compute Mean integrated squared error
    #======================================

    my_MISE[i] = np.sum(np.square(real_test_Y - fit_test_Y) * grid_width, 1).mean()
    np.savetxt('MISE_Epan_600size_samebw', my_MISE, newline='\n', delimiter = ' ', fmt="%s")
    
    print(i, ', time = ', time.time()-start_time)

#np.savetxt('MISE_Epam_2000size', my_MISE, newline='\n', delimiter = ' ', fmt="%s")
# np.savetxt('RMSPE', my_RMSPE, newline='\n', delimiter = ' ', fmt="%s")

```
